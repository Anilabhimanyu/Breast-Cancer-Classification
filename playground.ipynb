```python
# np.c_[bias_column, feature_1, feature_2]

import numpy as np

# 1. Suppose these are your raw 1D feature arrays
feature_1 = np.array([1.5, 2.3, 4.1])
feature_2 = np.array([10, 20, 30])

# 2. Add a column of ones (for the bias/intercept term)
# We use np.ones_like to match the length of our features
bias_column = np.ones_like(feature_1)

# 3. Use np.c_ to "zip" them into a 2D matrix (X)
# This replaces the need for .reshape(-1, 1) and np.hstack
X = np.c_[bias_column, feature_1, feature_2]

print("ML Input Matrix (X):")
print(X)
print("-" * 20)
print(f"Original shape: {feature_1.shape}")
print(f"New matrix shape: {X.shape}")
```

```python
# np.append(bias_column, feature_1, axis=0)

import numpy as np

# Existing feature matrix (2 samples, 2 features)
X = np.array([[25, 50000],
              [30, 60000]])

# New observation (1D array)
new_user = np.array([35, 70000])

# WRONG: np.append(X, new_user) -> returns [25, 50000, 30, 60000, 35, 70000]

# RIGHT: Reshape the new user to 2D and specify axis=0
X_updated = np.append(X, new_user.reshape(1, -1), axis=0)

print("Updated Matrix X:")
print(X_updated)
print(f"\nNew Shape: {X_updated.shape}")
```
